```python
import pandas as pd
```

```python
red=pd.read_csv("Red.csv.xls")
rose=pd.read_csv("Rose.csv.xls")
sparkling=pd.read_csv("Sparkling.csv.xls")
varieties=pd.read_csv("Varieties.csv.xls")
white=pd.read_csv("White.csv.xls")

df_list=["red","rose","sparkling","varieties","white"]
df_lists=[red,rose,sparkling,white]
```

```python
red["Type"]="red"
rose["Type"]="rose"
sparkling["Type"]="sparkling"
varieties["Type"]="varieties"
white["Type"]="white"
```

```python
wine_df=pd.concat(df_lists)
wine_df
```
