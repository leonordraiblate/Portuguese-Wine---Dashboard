```python
from bs4 import BeautifulSoup
import pandas as pd
soup = BeautifulSoup(response)
```

```python
winebox=[]
for product in soup.find_all("div",attrs={"data-product-tile-impression":True}):
    winebox.append(product.get("data-product-tile-impression").strip())
winebox
```

```python
import json
dict_list = [json.loads(j) for j in winebox]
df = pd.DataFrame(dict_list)
df
```

```python
df['category'] = df['category'].str.replace('Bebidas e Garrafeira/Vinhos/', '')
df
```

```python
#Vinho Tinto --> red
#Vinho Branco --> white
#Vinho do Porto --> porto
#Vinho Verde --> white
#Vinho Rosé --> rose
#Vinho Moscatel --> moscatel
#Sangrias e Aromatizados  --> sparkling 
#Vinho do Porto/Outros Vinhos do Porto --> porto
#Folhetos Pesquisa/Folheto Fim de Semana S52/Mais de 1000 Vinhos FDS S52 --> null
#Vinho Moscatel, da Madeira e Outros/Vinho da Madeira --> moscatel
```

```python
def combotype(row):
    if row["category"]=="Vinho Tinto":
        return "red"
    elif row["category"]=="Vinho Branco":
        return "white"
    elif row["category"]=="Vinho do Porto" or row["category"]=="Vinho do Porto/Outros Vinhos do Porto":
        return "red"
    elif row["category"]=="Vinho Verde":
        return "white"
    elif row["category"]=="Vinho Moscatel, da Madeira e Outros" or row["category"]=="Vinho Moscatel, da Madeira e Outros/Vinho da Madeira" :
        return "moscatel"
    elif row["category"]=="Sangrias e Aromatizados":
        return "sparkling"
    else:
        return "null"
    

df["Type"]= df.apply(combotype, axis = 1)
df
```

```python
df.to_excel('continente_df.xlsx', index=False)
```
